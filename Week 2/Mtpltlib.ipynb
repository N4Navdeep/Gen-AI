```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])
xpoints = np.array([2, 4, 6, 8])

plt.plot(ypoints, xpoints, marker = 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o:r')  #'r' for red colour and 'o' for line
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'green', mfc = 'r')  #ms = marker size, mec = marker edge colour, mfc = marker face colour
plt.show()
```

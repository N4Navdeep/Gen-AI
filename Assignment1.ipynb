#**Problem Statement:** Employee Salary Calculator

A company wants to calculate the net salary of its employees. Write a Python program that uses
functions to calculate the net salary based on the employee&#39;s basic salary.

**Requirements:**
* Create a function calculate_hra(basic_salary) that returns 20% of the basic salary.
* Create a function calculate_da(basic_salary) that returns 10% of the basic salary.
* Create a function calculate_net_salary(basic_salary) that calculates:

              
               Net Salary = Basic Salary + HRA + DA

* Accept the employee&#39;s name and basic salary as input.
* Display the salary breakdown and net salary.

```python
name=input("Enter employee name: ")
salary=int(input("Enter employee salary: "))
def calculate_hra():
  hra=salary*0.2
  return hra

def calculate_da():
  da=salary*0.1
  return da

def calculate_net_salary():
  har=calculate_hra()
  da=calculate_da()
  net=har+da+salary
  return net
print(calculate_net_salary())
```

#**Problem Statement:** Student Grade Management System

A school wants to automate the process of calculating students&#39; grades. Write a Python program
using functions to calculate the total marks, percentage, and final grade.

Requirements:

* Create a function calculate_total(marks) that returns the sum of marks.
* Create a function calculate_percentage(total, num_subjects) that calculates the
percentage.
* Create a function assign_grade(percentage) that returns:
    
     A for 90% and above

     B for 75–89%
    
     C for 60–74%
    
     D for 40–59%
    
     F for below 40%
* Accept marks for 5 subjects from the user.
* Display the total, percentage, and final grade.

```python
def calculate_total(marks):
    return sum(marks)

def calculate_percentage(total, num_subjects):
    return total / num_subjects

def assign_grade(percentage):
    if percentage >= 90:
        return "A"
    elif percentage >= 75:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 40:
        return "D"
    else:
        return "F"

marks = []

for i in range(1, 6):
    mark = float(input(f"Enter marks for Subject {i}: "))
    marks.append(mark)

total = calculate_total(marks)
percentage = calculate_percentage(total, len(marks))
grade = assign_grade(percentage)

print("Result: ")
print("Total Marks:", total)
print("Percentage:", percentage, "%")
print("Grade:", grade)
```

#**Problem Statement:** Student Performance Management System

A school wants to automate the process of managing student records. Develop a Python program
that stores student details and marks, calculates percentages, assigns grades, and displays the
topper.

**Requirements**

Your program should:

* Store student details (Roll No., Name, and Marks in 5 subjects).

* Use a dictionary to store student information.

* Store marks in a list.

* Store grade criteria in a tuple.

* Use functions to:

    o Calculate total marks.
    
    o Calculate percentage.
    
    o Assign grades.
    
    o Display student details.
* Use loops to:

    o Input data for multiple students.
    
    o Display all student records.

* Use control statements (if-elif-else) to assign grades:

    o A: 90% and above

    o B: 75–89%
    
    o C: 60–74%
    
    o D: 40–59%
    
    o F: Below 40%
* Display the student with the highest percentage.

```python
def grade(p):
    if p >= 90:
        return "A"
    elif p >= 75:
        return "B"
    elif p >= 60:
        return "C"
    elif p >= 40:
        return "D"
    return "F"

students = {}
n = int(input("Enter number of students: "))

for i in range(n):
    roll = input("Roll No: ")
    name = input("Name: ")

    marks = []
    for j in range(5):
        marks.append(int(input(f"Subject {j+1}: ")))

    total = sum(marks)
    per = total / 5

    students[roll] = {
        "name": name,
        "marks": marks,
        "percentage": per,
        "grade": grade(per)
    }

print("\nStudent Records")
for roll, data in students.items():
    print(roll, data["name"], data["percentage"], "%", data["grade"])

topper = max(students, key=lambda x: students[x]["percentage"])

print("\nTopper:")
print(students[topper]["name"], "-", students[topper]["percentage"], "%")
```

**Q1. Create a program that reads student marks from a file (filename.txt). Display the names of students who have passed (marks > 45) and those who have failed.**

Count the total number of passed and failed students.

Concepts Used:

* File Handling

* if-else

* for loop

* Counters

```python
file=open("/content/drive/MyDrive/GenAi/marks.txt","r")

passed = 0
failed = 0

for i in file:
  name, marks=i.strip().split(",")
  marks=int(marks)

  if marks>45:
    print(name,"Passed")
    passed+= 1
  else:
    print(name,"Failed")
    failed+= 1
file.close()
print("Total Passed:",passed)
print("Total Failed:",failed)
```

**Q2. A file (salary.txt) contains employee names and salaries. Read the file and display only those employees whose salary is greater than ₹50,000.** Also count how many employees satisfy this condition.

Concepts Used:

* File reading

* if statement

* Loop

* Counter

```python
f = open("/content/drive/MyDrive/GenAi/Salary.txt", "r")

count = 0

for i in f:
    name, salary = i.split()

    if int(salary) > 50000:
        print(name)
        count = count + 1

print("Total Employees =", count)

f.close()
```

**Q3. A file (attendance.txt) contains the attendance status of students
(Present/Absent).**

 Read the file and display only the names of absent students. Also
calculate the total number of present and absent students.

Concepts Used:

* File Handling

* if-else

* Loop

* Counting

```python
f = open("/content/drive/MyDrive/GenAi/Students.txt", "r")

present = 0
absent = 0

for i in f:
    name, status = i.split()

    if status == "Absent":
        print("Absent: ", name)
        absent = absent + 1
    else:
        present = present + 1

print("Total Present =", present)
print("Total Absent =", absent)

f.close()
```

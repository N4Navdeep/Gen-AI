```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_csv("/content/drive/MyDrive/GenAi/dataset1.csv")


def feature(msg):
    msg = msg.lower()
    spam = ["free","win","claim","offer","click","cash","prize","reward","lottery","urgent","iphone","recharge","exclusive"]
    ham = ["hi","call","meet","thank","birthday","morning","assignment","home","otp"]

    spam_count = 0
    ham_count = 0

    for word in spam:
        if word in msg:
            spam_count += 1

    for word in ham:
        if word in msg:
            ham_count += 1

    return [spam_count, ham_count, len(msg)]


X = []
for m in df["Message"]:
    X.append(feature(m))

y = df["Label"]

# Train KNN
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X, y)

# User input
text = input("Enter Message: ")

# Prediction
result = model.predict([feature(text)])

print("Prediction:", result[0])
```

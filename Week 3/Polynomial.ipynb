```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
df = pd.read_csv("/content/drive/MyDrive/GenAi/Position Salary.csv")
df
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
x = df[["Level"]]
y = df["Salary"]

model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y,label = 'Actual Data')
plt.plot(x,y_pred , label = 'Predicted Data')
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
x_poly = PolynomialFeatures(degree = 2)
```

```python
x_poly = x_poly.fit_transform(x)
```

```python
x_poly
```

```python
from sklearn.linear_model import LinearRegression
```

```python
poly_regressor = LinearRegression()
poly_regressor.fit(x_poly,y)
```

```python
new_pred=poly_regressor.predict(x_poly)
new_pred
```

```python
plt.scatter(x,y,label = 'Actual Data')
plt.plot(x,new_pred , label = 'Predicted Data')
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Polynomial Regression")
```

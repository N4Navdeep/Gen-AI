```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
X = [1, 2, 3, 4, 5, None, 7, 8, 9, 10, 11, 12, None, 14, 15, 16, 17, 18, 19, 20]
y_reg = [2, 4, 6, 8, 10, None, 14, 16, 18, 20, 22, 24, None, 28, 30, 32, 34, 36, 38, 40]
y_cls = [0, 0, 0, 0, 0, None, 0, 1, 1, 1, 1, 1, None, 1, 1, 1, 1, 1, 1, 1]
```

```python
df = pd.DataFrame({'X': X, 'y_reg': y_reg, 'y_cls': y_cls})
df
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace=True)
df
```

```python
plt.scatter(df['X'], df['y_reg'])
plt.title('Scatter Plot of X vs y_reg')
plt.xlabel('X')
plt.ylabel('y_reg')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['X']], df['y_reg'], test_size=0.2, random_state=42)
print(X_train)
```

```python
print(y_train)
```

```python
model = LinearRegression()
model.fit(df[['X']], df['y_reg'])
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

# Predict for X = 7
prediction_for_7 = model.predict([[7]])

print(f"The predicted output for X = 7 is: {prediction_for_7[0]:.2f}")
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(df[['X']], df['y_cls'])
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()
# Ensure y_cls is integer type for classification
df['y_cls'] = df['y_cls'].astype(int)
# Fit with the classification features and target
logistic_model.fit(df[['X']], df['y_cls'])
# Predict for X = 4 using the logistic model
prediction_for_4 = logistic_model.predict([[4]])

print(f"The predicted class for X = 4 is: {prediction_for_4[0]}")
```

```python
# KNN Classifier
import numpy as np
from sklearn.neighbors import KNeighborsClassifier


model = KNeighborsClassifier (n_neighbors=3)
model.fit(df[['X']], df['y_cls'])
prediction = model.predict([[5]])

print(prediction)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=10, random_state=50)
model.fit(df[['X']], df['y_cls'])
print(model.predict([[5]]))
```

```python
print ("Predictions of each models")
print(f"Linear Regression: {prediction_for_7[0]:.2f}")
print(f"Logistic Regression: {prediction_for_4[0]}")
print("KNN Classifier", prediction)
print("Random Forest", model.predict([[5]]))
```

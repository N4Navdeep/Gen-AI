```python
import numpy as np
import pandas as pd
import matplotlib as plt
from scipy import stats

full_health_data = pd.read_csv("/content/drive/MyDrive/GenAi/data29Jun.csv")
full_health_data = full_health_data.dropna()

x = pd.to_numeric(full_health_data["Average_Pulse"])
y = pd.to_numeric(full_health_data["Calorie_Burnage"])

slope, intercept, r, p, std_err = stats.linregress(x, y)
# slope =  slope (m) of the regression line
# intercept = intercept (c), i.e. , value of y when x = 0
# r = correlation coefficient (strength of relationship)
# p = P-value (checks if the relationship is statistically significant)
# std_err = standard error (accuracy of the calculted slope)
```

```python
import matplotlib.pyplot as plt

def myfunc(x):
  return slope *x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

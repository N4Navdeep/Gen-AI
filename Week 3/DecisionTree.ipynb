```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sns
```

```python
data = sns.load_dataset("iris")
data
```

```python
from sklearn.datasets import load_iris
```

```python
data1 = load_iris()
data1
```

```python
x = data1["data"]
y = data1["target"]
```

```python
y
```

```python
x
```

**Data Splition**

split the data in train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train, y_train
```

**Model Building**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(model, filled=True)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print("Testing acc: ", accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

**Post Pruning**

```python
model1 = DecisionTreeClassifier(max_depth=2)
```

```python
model1.fit(x_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(model1, filled=True)
```

**Prepruning Techinique**

```python
model2 = DecisionTreeClassifier()
```

```python
parameter = {
    "criterion": ["gini", "entropy"],
    "splitter": ["best", "random"],
    "max_depth": [1,2,3,4,5],
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
final_model = GridSearchCV(model2, param_grid=parameter, scoring="accuracy", cv=5)
final_model.fit(x_train, y_train)
```

```python

final_model.best_params_
```

```python
y_pred = final_model.predict(x_test)
```

```python
y_test, y_pred
```

```python
x_pred = final_model.predict(x_train)
```

```python
print("Testing acc: ", accuracy_score(y_test, y_pred))
print("Training acc: ", accuracy_score(y_train, x_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

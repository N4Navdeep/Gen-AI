```python
from sklearn.ensemble import RandomForestClassifier
```

```python
x = [[2], [3], [4], [5], [6], [7], [8]]
y = ["Fail", "Fail", "Fail", "Pass", "Pass", "Pass", "Pass"]

model = RandomForestClassifier(n_estimators=10, random_state=50)

model.fit(x, y)

print(model.predict([[5.5]]))
```

```python
x = [[120], [130], [140], [180], [190], [200]]
y = ["Apple", "Apple", "Apple", "Orange", "Orange", "Orange"]

model = RandomForestClassifier(n_estimators=10, random_state=42)

model.fit(x, y)

print(model.predict([[150]]))
```

#**Basic Random Forest**

```python
tree1 = "Cat"
tree2 = "Cat"
tree3 = "Dog"
tree4 = "Cat"
tree5 = "Cat"

votes = [tree1, tree2, tree3, tree4, tree5]

cat = votes.count("Cat")
dog = votes.count("Dog")

if cat > dog:
  print("Final Prediction: Cat")
else:
  print("Final Prediction: Dog")
```

#**Loop**

```python
trees = ["Cat", "Cat", "Dog", "Cat", "Cat", "Dog"]

cat = 0
dog = 0

for Prediction in trees:
  if Prediction == "Cat":
    cat += 1
  else:
    dog += 1

if cat > dog:
  print("Final Prediction: Cat")
else:
  print("Final Prediction: Dog")
```
